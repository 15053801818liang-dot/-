````python
import time
from IPython.display import clear_output


def monitor_audit_logs(log_path="workspace/logs/yushitai_audit.jsonl"):
    """实时监控御史台审计流（盘古推理日志）"""
    try:
        print("🔍 正在连接审计流... (按 Ctrl+C 停止)")
        while True:
            try:
                with open(log_path, "r", encoding="utf-8") as f:
                    lines = f.readlines()
                    tail = lines[-10:] if len(lines) >= 10 else lines

                clear_output(wait=True)
                print("═══════════════════════════════════════════════════════════")
                print("🧠 盘古 · 御史台实时审计监视窗")
                print(f"📂 源: {log_path}  |  总行数: {len(lines)}")
                print("═══════════════════════════════════════════════════════════\n")

                for line in tail:
                    print(line.strip())

            except FileNotFoundError:
                clear_output(wait=True)
                print("⏳ 等待审计日志生成... (请先运行 Pangu 推理任务)")

            time.sleep(1)

    except KeyboardInterrupt:
        print("\n🛑 监视器已停止。")


monitor_audit_logs()
````
